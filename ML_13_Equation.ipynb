```python
# Gradient Descent for y = (x + 3)^2

import matplotlib.pyplot as plt

# Define the function and its derivative
def f(x):
    return (x + 3)**2

def df(x):
    return 2 * (x + 3)

# Gradient Descent parameters
x = 2.0          # starting point changed to 2
learning_rate = 0.1
iterations = 30

# To store steps
x_vals = [x]
y_vals = [f(x)]

# Gradient Descent loop
for i in range(iterations):
    grad = df(x)
    x = x - learning_rate * grad
    x_vals.append(x)
    y_vals.append(f(x))
    print(f"Iteration {i+1:2d}: x = {x:.5f}, f(x) = {f(x):.5f}")

print("\nLocal minimum at x ≈", x)
print("f(x) ≈", f(x))

# Visualization
plt.figure(figsize=(7,5))
X = [i for i in range(-10,5)]
Y = [f(i) for i in X]
plt.plot(X, Y, label='y = (x + 3)^2')
plt.scatter(x_vals, y_vals, color='red', label='Gradient descent path')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('Gradient Descent Convergence (Starting from x = 2)')
plt.legend()
plt.grid(True)
plt.show()
```
